# Data scraping of Tables from The Browser using Selenium

### step1: import the requried libraries.

```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
```

### step2: Copy the url from required page source where the tables need to scrap and assign it to variable.

```python
url = 'https://en.wikipedia.org/wiki/List_of_mass_car_bombings'
```

### Step 3:- Set the path to chrome driver to configure web driver to use Chrome browser.

```python
browser = webdriver.Chrome()
```

### Step4:- Code to open the URL

```python
browser.get(url)
```

### Step5:- Read the HTML from page source

```python
html = browser.page_source
html
```

### Step 6: Parse HTML code and grab tables with Beautiful Soup

```python
soup = BeautifulSoup(html,'html.parser')
soup
```

### Step7:- Extracting the Table_data from the Page using select method

```python
table = soup.select('table.wikitable.sortable.jquery-tablesorter')
table
```

```python
table[0]
```

### Step8:- Extracting the Table _column names from table 

```python
thead = [i.text.strip() for i in table[0].thead.select('th')]
thead
```

### Step9:- Extracting the Table - rows Data from table

```python
trows = []
for r in table[0].tbody.select('tr'):
    trows.append([i.text.strip() for i in r.select('td') ])
```

```python
trows
```

### Step10:- Arranging the Columns and Rows in the Data Frame Formet.

```python
df = pd.DataFrame(trows)
df.columns = thead
df
```

### Step11:- Creating reusable code to scrap all the tables from the Page. 

```python
def MasCarBombing(table_no):
    thead = [i.text.strip() for i in table[0].thead.select('th')]
    trows = []
    for r in table_no.tbody.select('tr'):
        trows.append([i.text.strip() for i in r.select('td') ])
        df = pd.DataFrame(trows)
        df.columns = thead
    return df
```
